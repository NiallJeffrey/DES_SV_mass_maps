```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
import healpy as hp

plt.rcParams['figure.figsize'] = (15, 7)
plt.rcParams.update({'font.size': 18})
```

### 1) Kaiser-Squires with 10 arcminute Gaussian smoothing
### 2) Wiener filter reconstruction
### 3) GLIMPSE (sparse) reconstruction with $\lambda = 3.0$

(This requires unpacking "reconstructions.tar.gz")

```python
kaiser_squires_10am = np.load('reconstructions/published_des_sv_map_ks_10am.npy')
wiener = np.load('reconstructions/published_des_sv_map_wiener.npy')
glimpse = np.load('reconstructions/published_des_sv_map_glimpse_l3.npy')
```

```python
_ = plt.figure()
_ = hp.gnomview(kaiser_squires_10am, rot=[+75.0,-52.0],  title='Kaiser-Squires', \
                min = -0.03, max=0.03,  xsize=768, ysize=768, flip ='geo', cmap = 'inferno', sub = (1,3,1))
_ = hp.gnomview(wiener, rot=[+75.0,-52.0],  title='Wiener filter \n(Gaussian Prior)', \
                min = -0.03, max=0.03,  xsize=768, ysize=768, flip ='geo', cmap = 'inferno', sub = (1,3,2))
_ = hp.gnomview(glimpse, rot=[+75.0,-52.0],  title='GLIMPSE \n(Sparsity Prior)', \
                min = -0.03, max=0.03,  xsize=768, ysize=768, flip ='geo', cmap = 'inferno', sub = (1,3,3))
```
